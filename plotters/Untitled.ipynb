```python
import numpy as np
import matplotlib.pyplot as plt 

R= 0.5
N= 300
r1 = np.linspace(0, R, num=N)
t1 = np.linspace(-np.pi/4, np.pi/4, num=N)
r1, t1 = np.meshgrid(r1, t1)
x, y = r1*np.cos(t1) - r1*r1*r1*np.cos(3.*t1)/3., -r1*np.sin(t1) - r1*r1*r1*np.sin(3.*t1)/3.
r = np.sqrt(x**2 + y**2)
t = np.arctan2(y, x) 
plt.scatter(x.flatten(), y.flatten(), s=1)
```

```python
plt.scatter(r.flatten(), t.flatten(), s=0.1)
```
